```python
# Double digit numbers for training the model

import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import os

mnist = tf.keras.datasets.mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

def combine_images(image1, image2):
    resized_image1 = np.reshape(image1, (28, 28))
    resized_image2 = np.reshape(image2, (28, 28))
    combined_image = np.concatenate((resized_image1, resized_image2), axis=1)
    return combined_image

# Change the value of x accordingly. For example, if x=5, it will create images for the digits 50,51,...,59.
# Similarly if x=6, it will create images for the digits 60,61,..,69.
x=5
for i in range(0, 10):  
    x_indices_test = np.where(test_labels == x)[0]

    if x == i:
        total_combined_images = len(x_indices_test)
        print(f"Number of {x} images: {len(x_indices_test)}")
        print(f"Total number of combined images of {x}{i}: {len(x_indices_test)}")
        if not os.path.exists(f'Train/{x}{i}'):
                os.makedirs(f'Train/{x}{i}')

        for j, x_idx in enumerate(x_indices_test):
            mnist_image1 = test_images[x_idx]
            combined_image = combine_images(mnist_image1, mnist_image1)
            plt.imshow(combined_image, cmap='gray')
            plt.axis('off')
            plt.savefig(f'Train/{x}{i}/img_{j}.png', bbox_inches='tight', pad_inches=0)
            plt.close()

        print(f"{total_combined_images} combined images have been created and saved.")
        print()

    else:
        ith_indices_test = np.where(test_labels == i)[0]
        min_length = min(len(x_indices_test), len(ith_indices_test))

        total_combined_images_test = min_length

        print(f"Number of {x} images: {len(x_indices_test)}")
        print(f"Number of {i} images: {len(ith_indices_test)}")
        print(f"Total number of combined images of {x}{i}: {total_combined_images_test}")

        if not os.path.exists(f'Train/{x}{i}'):
            os.makedirs(f'Train/{x}{i}')

        for j in range(min_length):
            mnist_image1 = test_images[x_indices_test[j]]
            mnist_image2 = test_images[ith_indices_test[j]]
            combined_image = combine_images(mnist_image1, mnist_image2)

            plt.imshow(combined_image, cmap='gray')
            plt.axis('off')
            plt.savefig(f'Train/{x}{i}/img_{j}.png', bbox_inches='tight', pad_inches=0)
            plt.close()

        print(f"{total_combined_images_test} combined images have been created and saved.")
        print()
```
