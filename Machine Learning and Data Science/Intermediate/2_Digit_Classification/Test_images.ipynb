```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import os

mnist = tf.keras.datasets.mnist
(train_images, train_labels), (test_images, test_labels) = mnist.load_data()

def combine_images(image1, image2):
    resized_image1 = np.reshape(image1, (28, 28))
    resized_image2 = np.reshape(image2, (28, 28))
    combined_image = np.concatenate((resized_image1, resized_image2), axis=1)
    return combined_image

for a in range(10):
    x_indices_test = np.where(train_labels == a)[0][:10]
    total_images = len(x_indices_test)
    print(f"Total number of images of {a}: {total_images}")

    if not os.path.exists(f'Test/{a}'):
        os.makedirs(f'Test/{a}')
                
    for i, x_idx in enumerate(x_indices_test):
        mnist_image1 = train_images[x_idx]
        plt.imshow(mnist_image1, cmap='gray')
        plt.axis('off')
        plt.savefig(f'Test/{a}/img_{i}.png', bbox_inches='tight', pad_inches=0)
        plt.close()

    print(f"{total_images} images have been saved.\n")

for x in range(1, 10):
    for i in range(0, 10): 
        x_indices_test = np.where(train_labels == x)[0][:10]
        ith_indices_test = np.where(train_labels == i)[0][:10]
        total_combined_images = len(x_indices_test)

        if x == i:
            print(f"Total number of combined images of {x}{x}: {total_combined_images}")
        else:
            print(f"Total number of combined images of {x}{i}: {total_combined_images}")

        if not os.path.exists(f'Test/{x}{i}'):
            os.makedirs(f'Test/{x}{i}')

        for j, x_idx in enumerate(x_indices_test):  
            mnist_image1 = train_images[x_idx]
            ith_idx = ith_indices_test[j]
            mnist_image2 = train_images[ith_idx]
            combined_image = combine_images(mnist_image1, mnist_image2)

            plt.imshow(combined_image, cmap='gray')
            plt.axis('off')
            plt.savefig(f'Test/{x}{i}/img_{j}.png', bbox_inches='tight', pad_inches=0)
            plt.close()

        print(f"{total_combined_images} combined images have been created and saved.")
        print()
```
