```python
del matches["comp"]
del matches["notes"]
```

```python
matches["date"]=pd.to_datetime(matches["date"])#changing the type for convinience
```

```python
matches["venue_code"] = matches["venue"].astype("category").cat.codes#converting home/away into 0s and 1s
matches["opp_code"] = matches["opponent"].astype("category").cat.codes#generating a code for each team
matches["hour"] = matches["time"].str.replace(":.+", "", regex=True).astype("int")# 16:30 is converted to 16 using regular expression to search for it and converted to int
matches["day_code"] = matches["date"].dt.dayofweek#converting each day of the week to a no. like mon is 0, sun is 6
```

```python
matches
```

```python
matches["target"] = (matches["result"] == "W").astype("int")#win is 1 loss/draw is 0
```

```python
matches
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=150, min_samples_split=10, random_state=1)
#n_estimators=no of decision trees->the more the better
#min_samples_split=min no of samples in leaf of decision tree before splitting the node->if more less chance of overfitting but less accuracy on training data
#random_state=1->same result if data same
```

```python
train = matches[matches["date"] < '2024-01-01']#78% in training set(matches before 2024 in test set)
test = matches[matches["date"] > '2024-01-01']#22% in test set(matches in 2024 for test set)
```

```python
predictors = ["venue_code", "opp_code", "hour", "day_code"]
```

```python
rf.fit(train[predictors], train["target"])#fitting the model for training data
```

```python
preds = rf.predict(test[predictors])#for test data
```

```python
from sklearn.metrics import accuracy_score
accuracy= accuracy_score(test["target"], preds)#checking our accuaracy
accuracy
```

```python
combined = pd.DataFrame(dict(actual=test["target"], predicted=preds))#making a dataframe with result of test and train set
pd.crosstab(index=combined["actual"], columns=combined["predicted"])#returns a table telling how many times we were actually correct and how many times we were wrong
```

```python
#144 losses correctly predicted,62 losses wringly predicted
#50 wins correct,56 wins wrong
```

```python
from sklearn.metrics import precision_score
precision_score(test["target"], preds)#tells us how accurate our model is
```

```python
grouped_matches = matches.groupby("team")#groups by the team
```

```python
def rolling_averages(group, cols, new_cols):#func for calculating rolling averages
    group = group.sort_values("date")
    rolling_stats = group[cols].rolling(3, closed='left').mean()# means if for matchday4 dont include 4 calculate with 1,2,3
    group[new_cols] = rolling_stats
    group = group.dropna(subset=new_cols)#drop all missing values
    return group
```

```python
cols = ["gf", "ga", "sh", "sot", "dist", "fk", "pk", "pkatt"]
new_cols = [f"{c}_rolling" for c in cols]#gf_rolling,ga_rolinng,.....
matches_rolling = matches.groupby("team").apply(lambda x: rolling_averages(x, cols, new_cols))#applies lambda function to all the members in the data frame
```

```python
matches_rolling = matches_rolling.droplevel('team')#drop one level of index
```

```python
matches_rolling.index = range(matches_rolling.shape[0])#sets index to start from 0
```

```python
def make_predictions(data, predictors):#just making the last step into a function
    train = data[data["date"] < '2024-01-01']
    test = data[data["date"] > '2024-01-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)
    error = precision_score(test["target"], preds)
    return combined, error
```

```python
combined, error = make_predictions(matches_rolling, predictors + new_cols)#also taking in the new cols
```
