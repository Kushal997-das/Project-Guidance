Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from xgboost import XGBRegressor
from sklearn import metrics
```

```python
calories=pd.read_csv('/content/calories.csv')
```

```python
calories.head()
```

```python
excercise_data=pd.read_csv('/content/exercise.csv')
```

```python
excercise_data.head()
```

```python
## Merging the two dataframes

calories_data=pd.concat([excercise_data,calories['Calories']],axis=1)
```

```python
calories_data.head()
```

```python
## Finding out the shape of the dataframe
calories_data.shape
```

```python
calories_data.info()
```

```python
## Checking foe missing values
calories_data.isnull().sum()
```

```python
calories_data.replace({"Gender":{'male':0,'female':1}}, inplace=True)
```

```python
calories_data.head()
```

```python
## Separating features and target
X = calories_data.drop(columns=['User_ID','Calories'], axis=1)
Y = calories_data['Calories']
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

# **Model Training**

```python
model=XGBRegressor()
```

```python
model.fit(X_train, Y_train)
```

```python
test_data_prediction = model.predict(X_test)
```

```python
print(test_data_prediction)
```

**Mean Absolute Error**

```python
mae = metrics.mean_absolute_error(Y_test, test_data_prediction)
```

```python
print("Mean Absolute Error = ", mae)
```

The Error is very less hence it is a good model
