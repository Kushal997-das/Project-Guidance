```python
import numpy as np
import pandas as pd
import pickle
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
from matplotlib import pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('kaggle_diabetes.csv')
```

```python
df.head()
df.tail()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(x="Outcome", data=df, palette="bwr")
plt.show()
```

From the correlation plot we can see that glucose levels have a significant correlation with whether or no a patient has diabetes,another interesting conclusion we can draw is that skin thickness is correlated with BMI and insulin levels

```python
df_copy = df.copy(deep=True)
df_copy[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']] = df_copy[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0,np.NaN)
df.head()
```

```python
df.tail()
```

```python
df_copy['Glucose'].fillna(df_copy['Glucose'].mean(), inplace=True)
df_copy['BloodPressure'].fillna(df_copy['BloodPressure'].mean(), inplace=True)
df_copy['SkinThickness'].fillna(df_copy['SkinThickness'].median(), inplace=True)
df_copy['Insulin'].fillna(df_copy['Insulin'].median(), inplace=True)
df_copy['BMI'].fillna(df_copy['BMI'].median(), inplace=True)
```

```python
df.tail()
```

```python
X = df_copy[['Pregnancies',	'Glucose',	'BloodPressure',	'SkinThickness','Insulin',	'BMI',	'DiabetesPedigreeFunction',	'Age']]
```

```python
Y = df_copy[['Outcome']]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.20, random_state=0)
```

**Machine Learning Models**

*Logistic Regression*

Logistic Regression is similar to Linear Regression except it predicts if something is true or false, instead of fitting a line to the data like in case of Linear Regression, here an S shaped curve called the Sigmoid Function, although logistic regression gives you the probability of true or false it’s mostly used in binary classification problems, Now how this works is if the probability of it being true is greater than 50% it is true else it is false. It’s this ability to provide probabilities and classify samples using continuous and discrete measurements that makes it a popular classification algorithm, apart from binary classification there are many other types of logistic regression like multinomial logistic regression(many classes), ordinal logistic regression (multiple ordered classes).

```python
from sklearn.linear_model import LogisticRegression 
```

```python
clf =LogisticRegression(random_state=1).fit(X=X_train,y=y_train)
op=clf.predict(X_test)
clf.score(X_test,y_test)
```
